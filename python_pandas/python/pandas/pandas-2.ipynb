```python
df = pd.DataFrame(
    {
        "A" : ["foo","bar","foo","bar","foo","bar","foo","foo",],
        "B" : ["one","one","two","three","two","two","one","three"],
        "C" : np.random.randn(8),
        "D" : np.random.randn(8)
        
    }
)
df
```

```python
#"Group the DataFrame by the 'A' column and calculate the sum of the 'C' and 'D' columns for each group."
df.groupby("A")[["C", "D"]].sum()
```

```python
df.groupby(["A" ,"B"]).sum()
```

```python
#df2 will display the first 4 rows of the DataFrame, retaining the MultiIndex structure. Each row will have values 
# in columns "A" and "B" with their respective MultiIndex labels.
tuples = list(zip(*[["bar","bar","buz","buz","foo","foo","qux","qux"],
                    [ "one","two","one","two","one","two","one","two"]]))
index = pd.MultiIndex.from_tuples(tuples, names=["first", "second"])
df = pd.DataFrame(np.random.randn(8,2),index=index ,columns=["A","B"])
df2 = df[:4]
df2
```

```python
#stacked will be a Series where:
#The index now includes three levels: the original MultiIndex ("first" and "second") and the stacked column labels ("A" and "B").
#The values correspond to the values in columns "A" and "B" from the original DataFrame.
stacked = df.stack()
stacked
```

```python
stacked.unstack()
```

```python
stacked.unstack(2)
```
