```python
# import python libreries
import pandas as pd
import numpy as np
```

```python
# read the dayta from git hub raw url
url = "https://raw.githubusercontent.com/pandas-dev/pandas/refs/heads/main/pandas/tests/io/data/csv/tips.csv"
tips = pd.read_csv(url)
```

```python
# To see the top 5 rows of the file
# hear i am passing value as 8 ,if you are not passsing the value in arguments it will take the default value as 5
tips.head(8)
```

```python
# to select the required columns (pass the  column name in the form of strings)
tips[["total_bill", "tip", "smoker", "time"]]
```

```python
# generate new column with existing columns available data set
tips.assign(tip_rate=tips["tip"] / tips["total_bill"])
```

```python
# Check the pirticular value is available in the column (hear it is case sensitive)
is_dinner = tips["time"] == "Dinner"
is_dinner
```

```python
#  check the how many entries are available with selected value
is_dinner.value_counts()
```

```python
# getting the only true values of the selected value
tips[is_dinner]
```

```python
# filter the data based on multiple conditions (and)
# filter data time = dinner and tip > 5
df_dinner_and_tip_morethan_5 = tips[(tips["time"] == "Dinner") & (tips["tip"] > 5)]
df_dinner_and_tip_morethan_5
```

```python
# filter the data based on multiple conditions (or)
# filter size > 5 or total_bill > 45
df_size_morethan_5_and_total_bill_morethan_45 = tips[
    (tips["size"] > 5) | (tips["total_bill"] > 45)
]
df_size_morethan_5_and_total_bill_morethan_45
```

```python
# group by based on a column and knowing the size of the data set
tips.groupby("sex").size()
```

```python
# "Group data by a column and count non-null values in all columns."
tips.groupby("sex").count()
```

```python
# "Group data by a column and count non-null values in a specific column."
tips.groupby("sex")["tip"].count()
```

```python
# "Group data by a column and apply multiple aggregations to specific columns."
tips.groupby("day").agg({"day": np.size, "tip": np.mean})
```

```python
# "Group data by multiple columns and apply multiple aggregations to a specific column."
tips.groupby(["smoker", "day"]).agg({"tip": [np.size, np.mean]})
```

```python
# "Get the 5 smallest rows from the top 15 rows with the largest 'tip' values."
tips.nlargest(10 + 5, columns="tip").tail(5)
```

```python
# "Select the top 2 rows by 'total_bill' for each day and sort them by day and rank."
(
    tips.assign(
        rn=tips.sort_values(["total_bill"], ascending=False).groupby(["day"]).cumcount()
        + 1
    )
    .query("rn<3")
    .sort_values(["day", "rn"])
)
```

```python
# "Rank 'total_bill' within each day, select the top 2 for each day, and sort by day and rank."
(
    tips.assign(
        rnk=tips.groupby(["day"])["total_bill"].rank(method="first", ascending=False)
    )
    .query("rnk < 3")
    .sort_values(["day", "rnk"])
)
```
